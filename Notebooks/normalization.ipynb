```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/clean_data.csv')

df.head()
```

```python
X = df.copy()

url = X.pop('url')
label = X.pop('label')

y = pd.DataFrame().assign(url=url, label=label)

y
```

```python
X['starts_with_ip'] = X['starts_with_ip'].astype(int)
X['has_punycode'] = X['has_punycode'].astype(int)
X['domain_has_digits'] = X['domain_has_digits'].astype(int)
X['has_internal_links'] =X['has_internal_links'].astype(int)

X
```

```python
for col in X.select_dtypes("object"):
    X[col], _ = X[col].factorize()
    
discrete_features = X.dtypes == int

discrete_features
```

```python
norm_data = pd.concat([y, X], axis=1)

norm_data.head()
```

```python
norm_data.to_csv('../data/norm_data.csv', index=False)
```
